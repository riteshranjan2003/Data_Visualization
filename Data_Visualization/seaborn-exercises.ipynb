** *Note! In order to not lose the plot image, make sure you don't code in the cell that is directly above the plot, there is an extra cell above that one which won't overwrite that plot!* **

```python
# Ques 1: dataset-- titanic
    # Create the fare distribution through a histogram as shown below.
        ## output should be as shown below-
        
titanic = sns.load_dataset("titanic")
plt.hist(titanic[["fare"]], bins=30, ec="grey", color="pink")
plt.axis([0,600,0,500])
plt.xlabel("fare")
plt.show()
```

```python
# Ques 2: dataset-- titanic
   # Write a python program to create a boxplot for age and class group of all passengers.
    ## output should be as shown below-
    
sns.boxplot(data=titanic, x = titanic["class"], y=titanic['age'], palette="rainbow")
plt.grid(axis='y')
plt.show()
```

```python
# Ques 3: dataset-- titanic 
    # Create a bar chart for the number of male and female passengers.
        ## output should be as shown below-
        
bc = titanic.groupby("sex")["sex"].count().sort_values(ascending=False).plot(kind="bar", width=0.9, color=[(0.2,0.4,0.7), (0.3,0.5,0.3)])
bc.grid(axis='y', which="both")
print(bc)
```

```python
# Ques 4: dataset-- titanic
   # Set up a grid to plot survival probability against several variables like class, sex, who and alone.
   # Then Draw a seaborn pointplot onto each Axes
        ## output should be as shown below-
    
r=sns.PairGrid(titanic,x_vars=['class','sex','who','alone'],y_vars='survived', height=4.5, aspect=0.5)
r.map(sns.pointplot,color='#630330')
r.set(ylim=(0,1))
```

```python
# Ques 5: dataset-- diamonds
    # Generate the distribution of prices with respect to color for each cut quality.
    ## output should be as shown below
    
diamonds = sns.load_dataset("diamonds")
sns.barplot(data=diamonds, x = 'cut', y = 'price', hue='color', hue_order=['E', 'I', 'J', 'H', 'F', 'G', 'D'],
           order=['Ideal','Premium','Good','Very Good', 'Fair'])
plt.legend(ncol=4)
plt.xlabel("Cut",fontsize=15)
plt.ylabel("Price",fontsize=15)
plt.show()
```

```python
# Ques 6: dataset-- diamonds
   # obtain the counts of diamonds of each cut quality through bar chart.
    ## output should be as shown below
    
fig, ax = plt.subplots()
ax = sns.countplot(data=diamonds, x="cut", color="red", order=["Ideal", "Premium", "Good", "Very Good", "Fair"])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
print(ax)
```

```python
# Ques 7: dataset-- diamonds
# Create a histogram of the frequency of diamonds in the dataset with their respective carat specifications on the x-axis.
## output should be as shown below

sns.distplot(x=diamonds["carat"], kde= False)
```

```python
# Ques 8: dataset-- tips
    # Draw a nested violinplot and split the violins for easier comparison between day and total_bill.
    ## output should be as shown below
    
tips = sns.load_dataset("tips")
sns.set_style('whitegrid')
sns.violinplot(data=tips, x='day', y='total_bill',inner='quartile',palette=['#AECCE4','#FFFCBB'], hue='sex', split=True)
sns.despine(top=True, bottom=True, left=True, right=True)
```

```python
# Ques 9: dataset-- titanic
    # Draw a nested barplot to show survival for class and sex
    ## output should be as shown below

sns.barplot(data=titanic, x='class', y='survived', hue='sex')
plt.ylabel("survival probability")
plt.legend(title='sex', loc='center right', bbox_to_anchor=(1.2,0.5), frameon=False)
```

```python
# Ques 10: dataset-- fmri()
    # Create line plot to show the responses for different events and regions
    ## output should be as shown below

fmri = sns.load_dataset("fmri")
sns.lineplot(data=fmri, y='signal', x='timepoint', hue='region', style='event')
```

```python
# Ques 11: dataset-- iris
   # Create a scatter plot between sepal length and sepal width categorised by species.
            ## output should be as shown below

iris = sns.load_dataset("iris")
sns.set_style("ticks")
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
sns.despine(top=True, right=True)
plt.legend(loc="center right", frameon=False, bbox_to_anchor=(1.25,0.5))
```

```python
# Ques 12: dataset-- iris
   # Visualize the kernel density estimate of petal length for each species.
    ## output should be as shown below

for species in iris['species'].unique():
    sns.kdeplot(data=iris[iris['species']==species], x='petal_length', label=species)
sns.despine(top=True, right=True)
plt.ylabel("")
plt.legend(loc="center right", frameon=False, bbox_to_anchor=(1.25,0.5))
plt.show()
```

```python
# Ques 13: dataset-- iris
    # Study the relationship between sepal length and sepal width for the iris dataset by creating a joint plot.
    ## output should be as shown below
    
sns.jointplot(data=iris, x='sepal_length', y='sepal_width', marginal_kws={'color':"#6fa8dc"})
```

```python
# Ques 14: dataset-- iris
    # Create a paiplot for the dataset to show relation between each pair of features based on different species.
        ## output should be as shown below

sns.pairplot(data=iris, hue='species')
```

```python
# Ques 15: dataset-- exercise
   # Draw a pointplot to show pulse as a function of three categorical factors rest, walking and running.
    ## output should be as shown below

exercise=sns.load_dataset("exercise")
plt.figure(figsize=(5,4))
g = sns.FacetGrid(exercise, col='diet')
g.map(sns.pointplot, 'time', 'pulse', 'kind',hue_order=['rest', 'walking', 'running'])
plt.legend(frameon=False, loc="center right", bbox_to_anchor=(1.4,0.5))
plt.show()
```
